```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
train = pd.read_csv('./train.csv')   # (42000, 785)
train
```

```python
# 训练数据，目标
y = train['label']
X = train.drop('label', axis=1)
```

```python
a = np.array([[1,2,3],[2,3,4]])
np.sum(a)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from icecream import ic

pipe = Pipeline([
    ('std_scaler', StandardScaler()),
    ('clf', KNeighborsClassifier(weights='uniform'))
])

param_grid = {
    'clf__n_neighbors': [1],
    'clf__p': [1]
}

gsearch = GridSearchCV(estimator=pipe, param_grid=param_grid, cv=10, scoring='accuracy', n_jobs=-1, verbose=2)
gsearch.fit(X, y)
ic(gsearch.best_params_, gsearch.best_score_)
```
